
# Import Useful Libraries

```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn import preprocessing
```

# Loading Data and Arranging Training and Testing Sets


```python
data = pd.read_csv('concrete_data.csv')
```

```python
data.head()
```

```python
feature_labels = ['Cement', 'Blast Furnace Slag', 'Fly Ash', 'Water', 'Superplasticizer', 'Coarse Aggregate',
                  'Fine Aggregate', 'Age']
test_labels = ['Strength']
```

```python
feature_set = data[feature_labels]
test_set = data[test_labels]
feature_set = feature_set.to_numpy()
test_set = test_set.to_numpy()
inshape = feature_set[0].shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(feature_set, test_set, test_size=0.30)
```

```python
inshape = feature_set[0].shape
```

# PART-1

```python
def part_1():
    X_train, X_test, y_train, y_test = train_test_split(feature_set, test_set, test_size=0.30)
    model = Sequential([Dense(10, activation = 'relu', input_shape = inshape), Dense(1, activation = 'linear')])
    model.compile(optimizer = 'adam', loss = 'mse')
    model.fit(x = X_train, y = y_train, epochs = 50, verbose = False)
    MSE = model.evaluate(x = X_test, y = y_test, verbose = 0)
    return MSE
```

```python
num_iter = 50
MSE_matrix = np.zeros(num_iter)
for x in range(0, num_iter):
    MSE_matrix[x] = part_1()
```

```python
print('The Mean of Mean Squared Errors is : ',MSE_matrix.mean())
print('The Standard Deviation of Mean Squared Errors is : ',MSE_matrix.std())
```
